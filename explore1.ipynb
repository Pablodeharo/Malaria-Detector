Escribe un programa que imprima los N primeros números con las siguientes variantes:
Los múltiplos de tres, imprime "Tic" en lugar del número, y para los múltiplos de cinco, imprime "Toc". Para los números que son múltiplos tanto de tres como de cinco, imprime "TicToc".


Por ejemplo, para N = 101, el resultado sería: 

```python
def tic_toc(): # Definimos la funcion 
    for i in range(1, 102): # Iteramos con el bucle for sobre los numeros 
        if i % 3 == 0 and i % 5 == 0: # Modulo de multiplos de tres y cinoc (tictoc)
            print("Tic Toc")
        elif i % 3 == 0:
            print("Tic")
        elif i % 5 == 0:
            print("Toc")
        else:
            print(i) # imprimir numero si no cumple condición tic o toc

tic_toc() # llamar a la funcion
```
