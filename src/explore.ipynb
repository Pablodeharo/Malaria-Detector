```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import cv2
```

```python
# Define las rutas a las carpetas
train_dir = r'C:\Users\lenovo\Desktop\Malaria Detection\Malaria Detector\Data\Fotos\train'
parasitized_dir = os.path.join(train_dir, 'Parasitized')
uninfected_dir = os.path.join(train_dir, 'Uninfected')

# Contamos número de imágenes por carpeta
parasitized_count = len(os.listdir(parasitized_dir))
uninfected_count = len(os.listdir(uninfected_dir))

print(f'Número de fotos Parasitized: {parasitized_count}')
print(f'Número de fotos Uninfected: {uninfected_count}')

# Función para mostrar las primeras 25 imágenes de una categoría
def show_images(category_dir, category_name):
    images = os.listdir(category_dir)[:25]  # Obtener las primeras 25 imágenes
    plt.figure(figsize=(10, 10))
    for i, img_name in enumerate(images):
        img_path = os.path.join(category_dir, img_name)
        img = mpimg.imread(img_path)
        plt.subplot(5, 5, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.suptitle(f'Primeras 25 imágenes - {category_name}')
    plt.show()

# Mostrar las primeras 25 imágenes de cada categoría
show_images(parasitized_dir, 'Parasitized')
show_images(uninfected_dir, 'Uninfected')

# Función para cargar las imágenes en arrays de NumPy
def load_images(category_dir):
    images = []
    for img_name in os.listdir(category_dir)[:25]:  # Cambiar a None para cargar todas las imágenes
        img_path = os.path.join(category_dir, img_name)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img)
        else:
            print(f'Error al cargar la imagen: {img_path}')
    return images

# Cargar imágenes
parasitized_images = load_images(parasitized_dir)
uninfected_images = load_images(uninfected_dir)

# Función para obtener las dimensiones de las imágenes y los valores de los píxeles
def analyze_images(images, category_name):
    print(f"Análisis de imágenes - {category_name}")
    for idx, img in enumerate(images[:5]):  # Analizar las primeras 5 imágenes para no sobrecargar la salida
        print(f'Imagen {idx + 1} - Tamaño: {img.shape}')
        print(f'Imagen {idx + 1} - Valores de los primeros 5 píxeles:\n {img[:5, :5, :]}')
        plt.figure()
        plt.imshow(img)
        plt.title(f'Imagen {idx + 1}')
        plt.axis('off')
        plt.show()

# Analizar imágenes
analyze_images(parasitized_images, "Parasitized")
analyze_images(uninfected_images, "Uninfected")
```

Entrenamos el Modelo 


```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# Define las rutas a las carpetas
train_dir = r'C:\Users\lenovo\Desktop\Malaria Detection\Malaria Detector\Data\Fotos\train'
validation_dir = r'C:\Users\lenovo\Desktop\Malaria Detection\Malaria Detector\Data\Fotos\valid'
test_dir = r'C:\Users\lenovo\Desktop\Malaria Detection\Malaria Detector\Data\Fotos\test'
```

```python
# Crear generadores de datos
train_datagen = ImageDataGenerator(rescale=0.255)
validation_datagen = ImageDataGenerator(rescale=0.255)
test_datagen = ImageDataGenerator(rescale=0.255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(64, 64),
    batch_size=32,
    class_mode='binary'
)
```

```python
# Crear el modelo
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
```

```python
# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo
history = model.fit(                                                                      
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)                 
```

```python
# Evaluar el modelo en el conjunto de prueba
test_loss, test_accuracy = model.evaluate(test_generator, steps=test_generator.samples // test_generator.batch_size)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```

```python
# Graficar los resultados de entrenamiento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```

Resultado del entrenamiento

1. Inconsistencia en los resultados: hay epocas donde accuracy y los son 0, lo cual es inisual. Esto podría indicar un problema de flujo de datos.
2. Overfitting potencial: En algunas épocas, val accuracy llega a 1000, puede indicar sobreajuste
3. Fluctuaciones en val_ loss

Ajustes:

1. Regularización:
- Aumentar el dropout
- Añadir regularización L2 a las capas densas y convulucionales

```python
from tensorflow.keras.regularizers import l2
```

```python
# Crear el modelo
# Aumentamos dropout
# Añadimos regularización L2 a las capas 
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3), kernel_regularizer=l2(0.01)),  # Regularización L2
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.01)),  # Regularización L2
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.01)),  # Regularización L2
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu', kernel_regularizer=l2(0.01)),  # Regularización L2
    Dropout(0.7),  # Aumento de Dropout
    Dense(1, activation='sigmoid')
])
```

```python
# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Entrenar el modelo
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)
```

```python
# Evaluar el modelo en el conjunto de prueba
test_loss, test_accuracy = model.evaluate(test_generator, steps=test_generator.samples // test_generator.batch_size)
print(f'Test Loss: {test_loss}')
print(f'Test Accuracy: {test_accuracy}')
```

```python
# Graficar los resultados de entrenamiento
plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()
```
