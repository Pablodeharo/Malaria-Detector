```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import cv2
```

```python
# Define las rutas a las carpetas
train_dir = r'C:\Users\lenovo\Desktop\Malaria Detection\Malaria Detector\Data\Fotos\train'
parasitized_dir = os.path.join(train_dir, 'Parasitized')
uninfected_dir = os.path.join(train_dir, 'Uninfected')

# Contamos número de imágenes por carpeta
parasitized_count = len(os.listdir(parasitized_dir))
uninfected_count = len(os.listdir(uninfected_dir))

print(f'Número de fotos Parasitized: {parasitized_count}')
print(f'Número de fotos Uninfected: {uninfected_count}')

# Función para mostrar las primeras 25 imágenes de una categoría
def show_images(category_dir, category_name):
    images = os.listdir(category_dir)[:25]  # Obtener las primeras 25 imágenes
    plt.figure(figsize=(10, 10))
    for i, img_name in enumerate(images):
        img_path = os.path.join(category_dir, img_name)
        img = mpimg.imread(img_path)
        plt.subplot(5, 5, i + 1)
        plt.imshow(img)
        plt.axis('off')
    plt.suptitle(f'Primeras 25 imágenes - {category_name}')
    plt.show()

# Mostrar las primeras 25 imágenes de cada categoría
show_images(parasitized_dir, 'Parasitized')
show_images(uninfected_dir, 'Uninfected')

# Función para cargar las imágenes en arrays de NumPy
def load_images(category_dir):
    images = []
    for img_name in os.listdir(category_dir)[:25]:  # Cambiar a None para cargar todas las imágenes
        img_path = os.path.join(category_dir, img_name)
        img = cv2.imread(img_path)
        if img is not None:
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img)
        else:
            print(f'Error al cargar la imagen: {img_path}')
    return images

# Cargar imágenes
parasitized_images = load_images(parasitized_dir)
uninfected_images = load_images(uninfected_dir)

# Función para obtener las dimensiones de las imágenes y los valores de los píxeles
def analyze_images(images, category_name):
    print(f"Análisis de imágenes - {category_name}")
    for idx, img in enumerate(images[:5]):  # Analizar las primeras 5 imágenes para no sobrecargar la salida
        print(f'Imagen {idx + 1} - Tamaño: {img.shape}')
        print(f'Imagen {idx + 1} - Valores de los primeros 5 píxeles:\n {img[:5, :5, :]}')
        plt.figure()
        plt.imshow(img)
        plt.title(f'Imagen {idx + 1}')
        plt.axis('off')
        plt.show()

# Analizar imágenes
analyze_images(parasitized_images, "Parasitized")
analyze_images(uninfected_images, "Uninfected")
```
